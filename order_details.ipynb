```python
import pandas as pd
```

```python
df=pd.read_csv('orders.csv')
df
```

```python
## Handling null values
df.isnull().sum()
```

```python
df.fillna("Not available")
```

```python
#renaming columns
df.rename(columns={'Order Id': 'Order_id', 'Order Date':'Order_date', 'Ship Mode':'Ship_mode', 'Postal Code':'Postal_code', 'Sub Category':'Sub_category', 'Product Id':'Product_id', 'cost price':'Cost_price', 'List Price':'List_price', 'Discount Percent':'Discount_percent'}, inplace=True)
```

```python
df["Discount"]=df['List_price']*df['Discount_percent']/100
```

```python
df["Sale_price"]=df['List_price']-df['Discount']
```

```python
df["Profit"]=df['Sale_price']-df['Cost_price']
```

```python
df['Order_date']=pd.to_datetime(df['Order_date'], format="%Y-%m-%d")
```

```python
df.drop(columns={'Cost_price', 'List_price', 'Discount_percent'}, inplace=True)
```

```python
df.to_csv("order_detail.csv")
```
